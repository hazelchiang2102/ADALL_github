```python
import pandas as pd
import matplotlib.pyplot as plt

# Example: Replace this with the raw URL of your GitHub file
github_raw_url = 'https://raw.githubusercontent.com/hazelchiang2102/ADALL_github/refs/heads/main/laptop_prices_2024_sgd_TL.csv'
try:
    #This is a new change

  print("Loading data from GitHub...")
  df = pd.read_csv(github_raw_url)
  print("Successfully loaded data from GitHub!")
  display(df.head())

  #plot each column in df which is numeric using histogram
  print("Plotting each column in the DataFrame:")
  for column in df.columns:
    if pd.api.types.is_numeric_dtype(df[column]) and not pd.api.types.is_bool_dtype(df[column]):
      plt.hist(df[column], bins=10)
      plt.xlabel(column)
      plt.ylabel('Frequency')
      plt.title(f'Histogram of {column}')
      plt.show()
    elif pd.api.types.is_bool_dtype(df[column]):
      print(f"Skipping histogram for boolean column: {column}")

except Exception as e:
  print(f"Error loading data: {e}")
  print("Please ensure the URL is correct and the file format is compatible with `pd.read_csv`.")
```
